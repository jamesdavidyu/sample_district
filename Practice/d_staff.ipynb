```python
from matplotlib import pyplot as plt 
import pandas as pd
```

```python
d_staff=pd.read_csv('d_staff_answered.csv')
d_staff
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (5, 3)
plt.rcParams['figure.dpi'] = 75
```

```python
esperc=d_staff['enthusiastic_students'].value_counts(normalize=True)
esperc
```

```python
esperc=esperc.to_frame().reset_index()
esperc.columns=['response','percent']
esperc
```

```python
esperc['percent']=esperc['percent'].apply(lambda x: "{:.0f}%".format(x*100))
esperc
```

```python
plt.bar(esperc.response,esperc.percent,color='blue')
plt.show()
```
